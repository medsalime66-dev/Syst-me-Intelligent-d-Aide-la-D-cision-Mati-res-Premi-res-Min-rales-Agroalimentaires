## Interconnexion avec une base de données relationnelle (MySQL)

Dans cette étape nous récupérons des données stockées dans une base de données **MySQL**.

Actions réalisées :
- création d’une connexion avec SQLAlchemy
- exécution d’une requête SQL
- chargement du résultat dans un DataFrame Pandas
- analyse statistique des notes

```
import pandas as pd
from sqlalchemy import create_engine

# Création d'une connexion avec la base de données MySQL
engine = create_engine("mysql+pymysql://root:@localhost/school")

# Requête SQL pour sélectionner les étudiants ayant une note G3 > 15
query = "SELECT * FROM etudiants WHERE G3 > 15"

# stockage du résultat dans un DataFrame Pandas
df = pd.read_sql(query, engine)

# Affichage des 5 premières lignes
print("Apercu des etudiants(G3>15):")
display(df.head())

# Analyse statique des notes(G3)
print("les premieres statistiques descriptives des notes (G3):")
display(df.describe())
```

## Collecte de données non-structurées

Dans cette partie nous utilisons deux techniques d’acquisition de données en ligne :
- Web Scraping
- API

### Web Scraping

Le Web Scraping consiste à extraire automatiquement des informations depuis des pages web.

Nous utilisons les bibliothèques :
- `requests` pour envoyer des requêtes HTTP
- `BeautifulSoup` pour analyser le contenu HTML

```
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

Avant de scraper un site web, il est recommandé de vérifier le fichier **robots.txt** afin de connaître les règles d’exploration du site.

```
#verification du fichier robots.txt du site
url="https://books.toscrape.com/robots.txt"

#Envoi dune requete HTTP pour recuperer le fichier
reponse=requests.get(url)

#Affichage du contenu du fichier
print("Contenu du fichier:")
print(reponse.text)
```

Extraction des titres des livres présents sur la page principale du site.

```
#Url du site
url="https://books.toscrape.com"

#Envoi de la requete vers le site
reponse=requests.get(url)

#Recuperation du contenu HTML de la page
html=reponse.text

#Creation de l`objet BeautifulSoup pour parser le html
soup=BeautifulSoup(html,"html.parser")

#Recherche de toutes les balises <h3> qui contiennent les titres
titres=soup.find_all("h3")
#Creation d`une liste pour stocker les titres
liste_titres=[]

#parcours des resultats 
for t in titres:
    titre=t.text
    liste_titres.append(titre)

#Affichage des titres
print("Liste des titres:")
print(liste_titres)
```

Transformation des titres extraits en **DataFrame Pandas** pour faciliter leur analyse.

```
#Creation d`une DataFrame avec les titres
df=pd.DataFrame(liste_titres, columns=["titre"])

#Affichage des 5 premiers lignes
display(df.head())
```

### API

Les **APIs** permettent d’accéder à des données en ligne via des requêtes HTTP.

Dans cette étape nous récupérons des données depuis une API publique et nous les convertissons en DataFrame.

```
#url de l`API
api_url="https://jsonplaceholder.typicode.com/posts"

#Envoi de la requete vers l`API
reponse=requests.get(api_url)

#Conversation de la reponse en format JSON
data=reponse.json()

#Creation d`un DataFrame 
df=pd.DataFrame(data)

#Affichage des 5 premiers lignes
display(df.head())
```
